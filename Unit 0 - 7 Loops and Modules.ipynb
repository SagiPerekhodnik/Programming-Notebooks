## **`for` loops**
`for` loops are constructs that we can use to repeat code.

For example, let's print something out multiple times.

```python
# print five times
```

```python
 for i in range (1000):
        print(f"{i}. Mr Sagi is cooler!!!")
        
```

```python
# nested loop
for i in range (9):
    for j in range (9):
        print (f"{i} {j}")
```

```python
# print the numbers from 1 to 10
for i in range (100):
    print(i+1)
```

```python
# print numbers from 1 to 10, v2
# range ()
# range (<start>, <stop>, <>)
for i in range (1,11):
    print(i)
```

### Counting by numbers other than one
There are ways we can use `range` to count by numbers other than one.

```python
# count even numbers 2 to 10
# range (<start>, <stop-exc>, <step>)
for i in range (6, 23, 2):
    print(i)
```

```python
# count even numbers 2 to 10, v2
for i in range (5):
    print((i +4) / 8)
```

```python
# count backwards from 10 to 1
for i in range (50, 0, -3):
    print(i)
```

# **Lists and Loops**

## Lists

Lits are a type of **data**
* an ordered set of values 
* use `[]` to create our lists
* use `,` to separate individual elements

```python
# conventions
# --------
# name your lists as plural

# caluculate the avg %
percentages = [60, 34, 86, 67, 91, 37]

total = 0     # initializing a variable
for total in percentages: 
    total+=sum(percentages)


# for <element> in <list>:
for percentage in percentages:
    #  total = total + percentage
    total += percentages
    
print(f"The average is {total / len(percentages)}%")
```
